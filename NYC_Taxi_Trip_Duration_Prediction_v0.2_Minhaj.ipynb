# Feature Generation

## Extracting distance features from pickup and dropoff coordinates

### Calculating great circle distance

The great-circle distance (aka orthodromic distance or spherical distance) is the shortest distance between two points on the surface of a sphere, measured along the surface of the sphere. https://en.wikipedia.org/wiki/Great-circle_distance

```python
#Calculating great circle distance from pickup and dropoff coordinates
trip_df2=trip_df1.copy()
trip_df2['gc_distance']=trip_df2.apply(
    lambda x: great_circle(
        (x['pickup_latitude'], x['pickup_longitude']),
        (x['dropoff_latitude'], x['dropoff_longitude'])
    ).kilometers,
    axis=1)
```

```python
trip_df2.head()
```

### Calculating Manhattan Distance

The Manhattan distanceâ€™s name derives from the grid layout representing most streets on the island of Manhattan and the paths in it that give the shortest route between two points. Due to this, the Manhattan distance is also called the Taxicab distance or the City Block distance.
https://logicplum.com/knowledge-base/manhattan-distance/

```python
# Defining a function to calculate Manhattan distance
def manhattan_distance(pickup_lat, pickup_long, dropoff_lat, dropoff_long):
    distance = haversine((pickup_lat, pickup_long), (dropoff_lat, pickup_long)) + haversine((dropoff_lat, pickup_long), (dropoff_lat, dropoff_long))
    return distance
```

```python
#Calculating Manhattan distance from pickup and dropoff coordinates
trip_df2["mt_distance"] = trip_df2.apply(
    lambda x: manhattan_distance(
        x["pickup_latitude"], x["pickup_longitude"],
        x["dropoff_latitude"], x["dropoff_longitude"]), 
    axis=1)
```

## Feature Transformation

### Transforming Cyclical Features

Hours of the day, days of the week and months in a year are all examples that are cyclical in nature. Many new machine learning engineers donâ€™t think to convert these features into a representation that can preserve information such as hour 23 and hour 0 being close to each other and not far. The best way to handle this is to calculate the sin and cos component so that we can represent the cyclical feature as (x,y) coordinates of a circle. 
https://medium.com/ai%C2%B3-theory-practice-business/top-6-errors-novice-machine-learning-engineers-make-e82273d394db
http://blog.davidkaleko.com/feature-engineering-cyclical-features.html

In the given dataset, we have trip records only for 6 months i.e. from January to June 2016. Therefore, we'll transform only the day, hour and minute features into respective sin and cos components in this case. 

```python
# Calculating sine and cosine components of pickup_day, pickup_hr and pickup_min features.  
#trip_df4 = trip_df3.copy()
trip_df3['day_sin'] = np.sin((trip_df3.pickup_day)*(2.*np.pi/7))
trip_df3['day_cos'] = np.cos((trip_df3.pickup_day)*(2.*np.pi/7))
trip_df3['hr_sin'] = np.sin(trip_df3.pickup_hr*(2.*np.pi/24))
trip_df3['hr_cos'] = np.cos(trip_df3.pickup_hr*(2.*np.pi/24))
trip_df3['min_sin'] = np.sin(trip_df3.pickup_min*(2.*np.pi/60))
trip_df3['min_cos'] = np.cos(trip_df3.pickup_min*(2.*np.pi/60))
```

Now let's do a k-fold cross validation with the final model's hyper-parameters to get an idea of how the selected model generalizes over the entire dataset in terms of cross-val score. We'll use Root Mean Log Squared Error (RMSLE) metric as the score for cross-validation.  

```python
# Defining the scoring method to be used for cross validation
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer

def RMSLE(y, y_pred, **kwargs):
    out = np.sqrt(MSLE(np.expm1(y), np.expm1(y_pred)))
    return out

rmsle_scorer = make_scorer(RMSLE, greater_is_better=False)
```

As cross-validation using Random Forest with 100 trees will require a good amount of memory, we'll first clear some memory resources. 

```python
# Clearing memorry resources
import gc
del trip_df1
del trip_df2
del trip_df3
gc.collect()
```

```python
# Running 3-fold cross validation
rf4 = RandomForestRegressor(n_estimators=100, random_state =42)
scores = cross_val_score(rf4, X, y, cv=3, scoring=rmsle_scorer)
```

```python
abs(scores.mean())
```

```python
# Running 5-fold cross validation
rf4 = RandomForestRegressor(n_estimators=100, random_state =42)
scores = cross_val_score(rf4, X, y, cv=5, scoring=rmsle_scorer)
```

```python
abs(scores.mean())
```

As we can see, our final model has resulted in RMSLE equal to 0.312 on the test dataset. And when 3-fold and 5-fold cross validations have been performed, it has resulted in RMSLE scores equal to 0.312 and 0.310 respectively. So we can say that the final model generalizes well.
